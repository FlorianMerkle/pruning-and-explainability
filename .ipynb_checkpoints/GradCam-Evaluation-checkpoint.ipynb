```python
from pathlib import Path
import torch
import matplotlib.pyplot as plt
import pandas as pd

if torch.cuda.is_available() == True:
    device = torch.device("cuda:0")
else:
    device = torch.device("cpu")
print(device)
dtype = torch.float32
```

# Load Images

```python
IMGS_PATH = Path('./exp-prune-data/imagelist.pt')
LABELS_PATH = Path('./exp-prune-data/labellist.pt')
img_batch = torch.stack(torch.load(IMGS_PATH)).to(device)
label_batch = torch.stack(torch.load(LABELS_PATH)).to(device)
```

# Load Heatmaps

```python
LR_PATH = './gradcam_hms/low_res/'
HR_PATH = './gradcam_hms/high_res/'
ARCHITECTURE = 'VGG16'
CPRS = ['CPR1', 'CPR2', 'CPR4', 'CPR8', 'CPR16', 'CPR32', 'CPR64' ]
lr_hms = dict()
hr_hms = dict()
for CPR in CPRS:
    lr_hms[CPR] = torch.load(Path(LR_PATH,f'{CPR}-{ARCHITECTURE}-lr_heatmaps.pt'))
    hr_hms[CPR] = torch.load(Path(HR_PATH,f'{CPR}-{ARCHITECTURE}-hr_heatmaps.pt'))
```

```python
fig, axes = plt.subplots(len(CPRS),5, figsize=(15,30))
for i, CPR in enumerate(CPRS):
    for j in range(5):
        axes[i,j].imshow(img_batch[j].detach().cpu().permute(1,2,0))
        axes[i,j].imshow(hr_hms[CPR][j], cmap='jet', alpha=.4)
        if j==2:axes[i,j].set_title(CPR)
```

```python
fig, axes = plt.subplots(len(CPRS),5, figsize=(20,30))
for i, CPR in enumerate(CPRS):
    for j in range(5):
        axes[i,j].imshow(lr_hms[CPR][j], cmap='jet')
        if j==2:axes[i,j].set_title(CPR)
```

# Pixel Stats

```python
# mean value of the low-res gradcam heatmaps
mean_px_value = pd.Series([lr_hms[CPR].mean(dim=(1,2)).mean().item() for CPR in CPRS], name='mean', index=CPRS)
# mean variance (per low-res gradcam heatmap)
px_var = pd.Series([lr_hms[CPR].var(dim=(1,2)).mean().item() for CPR in CPRS], name='variance', index=CPRS)
# mean non_zero values per low-res gradcam heatmap
non_zeros = pd.Series([lr_hms[CPR].count_nonzero().item()/len(lr_hms[CPR]) for CPR in CPRS], name='non-zero', index=CPRS)
```

```python
pd.DataFrame([mean_px_value,px_var,non_zeros])
```

```python
for i, CPR1 in enumerate(CPRS):
    for CPR2 in CPRS[i+1:]:
        print('L_1 Distance between',CPR1,'&', CPR2,':', round(((lr_hms[CPR1] - lr_hms[CPR2]).norm(p=1)/img_batch.shape[0]).item(),2))
    print(10*'----')
```
